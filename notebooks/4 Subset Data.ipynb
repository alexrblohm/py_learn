```python
#basic data-related pyton commands

#load pandas module
import pandas as pd

#read in tree data
tree = pd.read_csv('../data/TreeWeight.csv')

#.head, .describe are methods
print(tree.head(10))
tree.describe()

#select a column 

#this way works if the column name does not have any spaces or special characters
tree_var1 = tree.RootStock
print(tree_var1)

#another way
tree_var2 = tree['RootStock']
print(tree_var2)

```

```python
#select more than one by name
sub_tree1 = tree[['RootStock','TrunkGirth']]
print(sub_tree1)

#indexing in python is WEIRD
#starts at zero, end point is NOT included
#this gives columns 1 and 2
sub_tree2 = tree[tree.columns[0:2]]
print(sub_tree2)
```

```python
#using the loc and iloc methods

#select all rows, certain columns
#loc requires the name of the column

loc_meth = tree.loc[:,'RootStock':'TrunkGirth']
print(loc_meth)

#same thing using iloc, which requires the index

iloc_meth = tree.iloc[:, 0:2]
print(iloc_meth)

#select particular rows and columns
#row argument is the same for loc and iloc
row_col_tree = tree.iloc[0:10, 0:2]
print(row_col_tree)
```

```python
#subset based on condition

#logical statements return T/F
#use these to select rows
#the . is like $ in R in this case

lil_girth = tree[tree.TrunkGirth < 350]
print(lil_girth)

stock_x = tree[tree.RootStock == 'X']
print(stock_x)

#same result as above
tree[tree['RootStock'] == 'X']
```
